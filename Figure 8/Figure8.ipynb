## Plot/save mRNA concentration dynamic for weak/strong oscillators and selected FRPs, with average line for activity onset visualization

```python
"""
Model of circadian oscillator with entraiment
"""
import numpy as np
import matplotlib as plt
from scipy.integrate import odeint
from pylab import *
from scipy import signal
from scipy.signal import argrelextrema
import pandas as pd
import seaborn as sns


def hill_up(x, k, n):
    h = np.power(x,n)/(np.power(k,n) + np.power(x,n))
    return h
    
def hill_down(x, k, n):
    h = np.power(k,n)/(np.power(k,n) + np.power(x,n))
    return h

def square_wave(t, period, amp):
    wave = amp*signal.square(2 * np.pi * t/period)+amp
    return wave
            
# solve the system dy/dt = f(y, t)
def df(y, t):
    #assign input vector,y to variables of the model
    M  = y[0]
    PC = y[1]
    PN = y[2]

    # mRNA expression hill_down(x, k, n)
    dM =  vs*hill_down(PN, KI, 4) - vm*hill_up(M, Km, 1) - square_wave(t, light_period, light_amp)*M
    # Protein concentration, PC - celllular, PN - nuclear
    dPC =   ks*M + k2*PN - k1*PC    - vd*hill_up(PC, Kd, 1)
    dPN =        - k2*PN + k1*PC
    # output right-hand side of differetial equations
    return [dM, dPC, dPN]

def find_avg_crossing(t,v):
    # convert voltage trace into array
    v_array = np.array(v, dtype=float)
    v_avg = np.mean(v)
    # find indices of mean crossings
    crossings = []
    for i in range(0,len(v)-1):
        if (v[i]<v_avg) and (v[i+1]>v_avg):
            crossings.append(i)
    # Record times and voltages at mean crossings
    print('crossings = ',crossings)
    crossing_times = t[crossings]
    crossing_amps = v[crossings]
    print('crossing_times = ',crossing_times)
    return(crossing_times, crossing_amps)


#-----------------------------------------------------------------------------------------------
#   Make a dataframe with oscillator description, such as type and color it would be plotted
#    The index of the dataframe contain parameter vaule of an oscillator
#-----------------------------------------------------------------------------------------------
def oscillator_summary():
    vs_list   = [1.0,  1.6   ]
    vs_names  = ['Weak', 'Strong']
    vs_colors = ['b', 'r']
    df = pd.DataFrame(zip(vs_names, vs_colors), columns =[ 'type', 'color'], index = vs_list)
    return(df)

# Choose total time of simulation, T
T = 552.
# Chose trainsients time
T_transients = 480
T_plot = T - T_transients

# Chose time step
dt=0.05

#Calculate number of time steps
nsteps=int(T/dt)

# Calculate number of indices in transient part
trans_ind = int(T_transients/dt)

# Create time grid
t  = np.linspace(0, T, nsteps) 


# Choose initial conditions
M0  = 0.89009363    
PC0 = 0.83699889
PN0 = 1.25723695
y0= [2.1740351,  3.22519908, 3.13304933]  

# Paramters of the model
#vs = 1.6
vm = 0.505
Km = 0.5
KI = 1.0
#ks = 0.3
vd = 1.4
Kd = 0.13
k1 = 0.5
k2 = 0.6

vs_df = oscillator_summary()
light_amp_list    = [2.0]
light_period = 24

count = int(T_plot/light_period)
max_val = 15

#frp_choice = 18
#frp_choice = 22
#frp_choice = 26
frp_choice_list = [18,24,30]

for vs in vs_df.index:
    df_protocol= pd.read_csv(vs_df.type[vs]+'_periods_ks.csv', index_col = 0)
    for ind in df_protocol.index: 
        ks  = df_protocol.ks[ind]
        entr_period = df_protocol.periods[ind]
        print(ind,entr_period,ks)
        for light_amp in light_amp_list:
            # Solve ODE and plot mRNA concentraion, M
            print('vs, ks, light_amp = ',vs,ks,light_amp)
            soln = odeint(df, y0, t)
            M  = soln[trans_ind:, 0]
            PC = soln[trans_ind:, 1]            
            time = t[trans_ind:]-t[trans_ind]
            [crossings, amps] = find_avg_crossing(time,M)
            print('crossings: ',crossings)
            print('amplitudes: ',amps)
            mrna_onset_times = crossings - 60.    # choose a specific onset of dark phase to compare to
            lo = 0
            while (mrna_onset_times[lo]<(-0.1)):    # find first delay time relative to this specific onset
                lo=lo+1
            if (ind in frp_choice_list):                
                plt.figure(figsize=(10,4))
                plt.plot(time, M, vs_df.color[vs], linewidth=4)
                # plot a dotted line at average value of M
                M_array = np.array(M, dtype=float)
                M_avg = np.mean(M_array)
                M_avg_plot = [M_avg for i in range(len(time))]
                plt.plot(time, M_avg_plot, 'k--', linewidth = 4)                                   
                for j in range(0,count):
                    inc=j*light_period
                    plt.fill([12+inc,12+inc,24+inc,24+inc],[0,max_val,max_val,0],'k', alpha=0.2, edgecolor='k')
                plt.xlabel("Time (hr)",fontsize=20)
                plt.ylabel("[M] (nM)",fontsize=20)
                plt.xticks(fontsize=16)
                plt.yticks(fontsize=16)
                plt.xlim(0,T_plot)
                plt.ylim(0,max_val)
                ax = plt.gca()
                xticks = ax.xaxis.get_major_ticks() 
                xticks[0].label1.set_visible(False)
                ax.spines['right'].set_color('none')
                ax.spines['top'].set_color('none')
                title_string = vs_df.type[vs] + ' Oscillator, FRP = ' + str(ind)+ ' hr'
                plt.title(title_string,fontsize=24)
                plt.tight_layout()
                filename = "Onset-P"+str(ind)+"-"+vs_df.type[vs]+".jpeg"
                plt.savefig(filename, dpi = 300)
                plt.show()             
```

## Create plot of activity onset vs FRP for weak and strong oscillators

```python
import seaborn as sns
import numpy as np
import matplotlib as plt
from scipy.integrate import odeint
from pylab import *
from scipy import signal
from scipy.signal import argrelextrema
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

fig, ax = plt.subplots()

def hill_up(x, k, n):
    h = np.power(x,n)/(np.power(k,n) + np.power(x,n))
    return h
    
def hill_down(x, k, n):
    h = np.power(k,n)/(np.power(k,n) + np.power(x,n))
    return h

def square_wave(t, period, amp):
    wave = amp*signal.square(2 * np.pi * t/period)+amp
    return wave
            
# solve the system dy/dt = f(y, t)
def df(y, t):
    #assign input vector,y to variables of the model
    M  = y[0]
    PC = y[1]
    PN = y[2]

    # mRNA expression hill_down(x, k, n)
    dM =  vs*hill_down(PN, KI, 4) - vm*hill_up(M, Km, 1) - square_wave(t, light_period, light_amp)*M
    # Protein concentration, PC - celllular, PN - nuclear
    dPC =   ks*M + k2*PN - k1*PC    - vd*hill_up(PC, Kd, 1)
    dPN =        - k2*PN + k1*PC
    # output right-hand side of differetial equations
    return [dM, dPC, dPN]

#-----------------------------------------------------------------------------------------------
#   Make a dataframe with oscillator description, such as type and color it would be plotted
#    The index of the dataframe contain parameter vaule of an oscillator
#-----------------------------------------------------------------------------------------------
def oscillator_summary():
    vs_list   = [1.0 , 1.6   ]
    vs_names  = ['Weak', 'Strong']
    vs_colors = ['b', 'r']
    df = pd.DataFrame(zip(vs_names, vs_colors), columns =[ 'type', 'color'], index = vs_list)
    return(df)

# Choose total time of simulation, T
T = 600.
# Chose trainsients time
T_transients = 480
T_plot = T - T_transients

# Chose time step
dt=0.05

#Calculate number of time steps
nsteps=int(T/dt)

# Calculate number of indices in transient part
trans_ind = int(T_transients/dt)

# Create time grid
t  = np.linspace(0, T, nsteps) 

# Choose initial conditions
M0  = 0.89009363    
PC0 = 0.83699889
PN0 = 1.25723695
y0= [2.1740351,  3.22519908, 3.13304933]  

# Paramters of the model

vm = 0.505
Km = 0.5
KI = 1.0
#ks = 0.3
vd = 1.4
Kd = 0.13
k1 = 0.5
k2 = 0.6

vs_df = oscillator_summary()
light_amp = 2.0
light_period = 24
#entrain_period = 25

entrain_period = list(range(18,31))

f_entrain_period = pd.DataFrame(entrain_period, columns =['frp']) 

count = int(T_plot/light_period)  # number of light periods in plotting region

for vs in vs_df.index:
    print('vs = ',vs)
    df_protocol= pd.read_csv(vs_df.type[vs]+'_periods_ks.csv', index_col = 0)
    onset = []
    for column in f_entrain_period:
        columnSeriesObj = f_entrain_period[column]
        ks_series = df_protocol.ks[columnSeriesObj.values]
        
    for n in entrain_period:
        ks = ks_series[n]
        soln = odeint(df, y0, t)
        M  = soln[trans_ind:, 0]
        M_avg = np.mean(M)
        time = t[trans_ind:]-t[trans_ind]  
        threshold = M_avg
        crossings = []
        for i in range(0,len(M)-1):
            if (M[i]<threshold) and (M[i+1]>threshold):
                crossings.append(i)
        d_period = 12 
        onset.append(t[crossings[0]] - d_period)   
    data = list(zip(entrain_period, onset))    
    df_p = pd.DataFrame(data, columns = ['onset', 'entrain_period'])
    df_p.to_csv(vs_df.type[vs]+'_onset_frp.csv', index=True)
    ax = sns.lineplot(entrain_period, onset, color = vs_df.color[vs], label = vs_df.type[vs])
    ax.set_ylabel('Activity onset (hr)',  fontsize=11.5, fontname= 'Arial')
    ax.set_xlabel('FRP (hr)',   fontsize=11.5,  fontname= 'Arial')
    ax = plt.gca()
    xticks = ax.xaxis.get_major_ticks() 
    xticks[0].label1.set_visible(False)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
plt.legend(loc = 'best')    
plt.savefig('Activity-onset-vs-frp.jpeg',dpi = 300)
plt.show()
```
